```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
```

```python
data_dic = {
    'title' : [],
    'con' : []
}
```

```python
for i in range(1,618) :
    time.sleep(0.001)
    url = f'http://premium.etnews.com/ict/index.html?id={i}'
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html5lib')
    if source.find('script').text=="alert('존재하지 않는 컨텐츠입니다.');history.go(-1)":
        i=i+1
        continue
    title = source.find('h5')
    con = source.find('div',{'class':'con '})
    data_dic['title'].append(title.get_text().replace("  ","").rstrip())
    data_dic['con'].append(con.get_text().replace("  ","").replace("\n",""))
    print(f'{i}/617')
```
